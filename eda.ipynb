```python
# plot a histogram of amounts and color code by class label
def plot_feature_histogram_by_class(df: pd.DataFrame, feature: str, target: str) -> None:
    """
    Plots overlaid histograms of a feature for the positive and negative classes.

    Args:
        df (pd.DataFrame): The input DataFrame.
        feature (str): The name of the feature to plot.
        target (str): The name of the target column with binary class labels (0 or 1).
    """
    positive = df[df[target] == 1][feature]
    negative = df[df[target] == 0][feature]

    plt.figure(figsize=(10, 6))
    plt.hist(positive, bins=150, alpha=0.7, label="Positive Class (1)", color='skyblue', density=True)
    plt.hist(negative, bins=1500, alpha=0.7, label="Negative Class (0)", color='salmon', density=True)
    # plt.yscale('log')
    plt.xlabel(feature)
    plt.ylabel("Density")
    plt.title(f"Histogram of '{feature}' by Class")
    plt.xlim(0, 1500)
    plt.ylim(0, 0.015)
    plt.legend()
    plt.grid(True)
    plt.show()

plot_feature_histogram_by_class(df, 'Amount', target = 'Class')
# df['Class'].values
```

```python
# plot a histogram of amounts and color code by class label
def plot_feature_histogram_by_class(df: pd.DataFrame, feature: str, target: str) -> None:
    """
    Plots overlaid histograms of a feature for the positive and negative classes.

    Args:
        df (pd.DataFrame): The input DataFrame.
        feature (str): The name of the feature to plot.
        target (str): The name of the target column with binary class labels (0 or 1).
    """
    positive = df[df[target] == 1][feature]
    negative = df[df[target] == 0][feature]

    plt.figure(figsize=(10, 6))
    plt.hist(positive, bins=75, alpha=0.7, label="Positive Class (1)", color='skyblue', density=True)
    plt.hist(negative, bins=1500, alpha=0.7, label="Negative Class (0)", color='salmon', density=True)
    # plt.yscale('log')
    plt.xlabel(feature)
    plt.ylabel("Density")
    plt.title(f"Histogram of '{feature}' by Class")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_feature_histogram_by_class(df, 'Time', target = 'Class')
```
